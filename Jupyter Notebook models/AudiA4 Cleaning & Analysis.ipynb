```python
def convert_color(color):
    if color in ['Zlatna','Kameleon', 'Smeđa', "Bež","Tirkiz","Ljubičasta","Bordo","Narandžasta","Braon","Crvena", "Zelena"]:
        color="Drugo"
        return color
    else:
        return color
df['Boja'] = df['Boja'].apply(lambda x: convert_color(x))
```

```python
def convert_vol(volume):
    if volume<=1410 and volume>=1350:
        volume=1.4
        return volume
    elif volume<=1610 and volume>=1550:
        volume=1.6
        return volume
    elif volume<=1810 and volume>=1750:
        volume=1.8
        return volume
    elif volume<=1910 and volume>=1850:
        volume=1.9
        return volume
    elif volume<=2010 and volume>=1950:
        volume=2
        return volume
    elif volume<=2210 and volume>=2150:
        volume=2.2
        return volume
    elif volume<=2310 and volume>=2250:
        volume=2.3
        return volume
    elif volume<=2510 and volume>=2450:
        volume=2.5
        return volume
    elif volume<=3010 and volume>=2950:
        volume=3.0
        return volume
    

df['Kubikaza'] = df['Kubikaza'].apply(lambda kub: convert_vol(kub))
```

```python
sns.countplot(df['Kubikaza'], data=df)
```
